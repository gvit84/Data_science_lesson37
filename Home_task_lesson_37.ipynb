```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.linear_model import Ridge
from sklearn.linear_model import Lasso
from sklearn.linear_model import ElasticNet
from sklearn.model_selection import GridSearchCV
```

Нормалізуємо наш датасет, шляхом видалення строк, в яких відсутнє хоча б одне значення:

```python
data = pd.read_csv('C:/Users/Vitalii/Downloads/communities.csv', header=None, na_values=['?'])
data
```

```python
data2 = data.dropna()
data2
```

```python
X = data2.drop(columns = [127, 3]).values
y = data2[127].values
```

```python
X, y
```

Створимо тренувальний та тестовий датасети та побудуємо модель поліноміальної регресії:

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.30, random_state = 1)
```

```python
lr = LinearRegression().fit(X_train, y_train)
```

```python
print(lr.score(X_train, y_train))
print(lr.score(X_test, y_test))
```

Виконаємо регуляризації L1 (Lasso), L2 (Ridge) та ElasticNet (комбінація Lasso та Ridge регуляризацій):

```python
lasso = Lasso(alpha=0.002).fit(X_train, y_train)
```

```python
print(lasso.score(X_train, y_train))
print(lasso.score(X_test, y_test))
```

```python
ridge = Ridge(alpha=0.95)
ridge.fit(X_train, y_train)
```

```python
print(ridge.score(X_train, y_train))
print(ridge.score(X_test, y_test))
```

```python
elastic_net = ElasticNet(alpha=0.005, l1_ratio=0.5).fit(X_train, y_train)
```

```python
print(elastic_net.score(X_train, y_train))
print(elastic_net.score(X_test, y_test))
```

В результаті для даного набору даних найкращим методом регуляризації виявився метод "Lasso".
